```python
# Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Find and display duplicate rows
duplicates = df[df.duplicated(keep=False)]

# Print the duplicate rows
print("Duplicate Rows:")
print(duplicates)
```

```python
df_no_duplicates = df.drop_duplicates(keep='last')
```

```python
# Checking if there are any duplicates in the data
df_no_duplicates.duplicated().sum()
```

```python
df_no_duplicates.info()
```

```python
df_no_duplicates.head()
```

```python
# checking for different types of age groups
df_no_duplicates['AgeGroup'].unique()
```

```python
# Create a mapping dictionary for age groups
age_group_mapping = {
    '18-25': 1,
    '26-35': 2,
    '36-45': 3,
    '46-55': 4,
    '55+': 5
}

# Map the values in the 'AgeGroup' column to numerical values using .loc
df_no_duplicates.loc[:, 'AgeGroup'] = df_no_duplicates['AgeGroup'].map(age_group_mapping)
```

```python
df_no_duplicates.head()
```

```python
df_no_duplicates['AgeGroup'].unique()
```

```python
# Assuming your DataFrame is named 'df_no_duplicates'
df_no_duplicates.loc[:, 'AgeGroup'] = df_no_duplicates['AgeGroup'].astype(int)
```

```python
df_no_duplicates.info()
```

```python
# Set the maximum number of displayed columns to None
pd.set_option('display.max_columns', None)
```

```python
df_no_duplicates.head()
```

```python
df_no_duplicates = df_no_duplicates.dropna(axis=0)
```

```python
df_no_duplicates = df_no_duplicates.reset_index(drop=True)
```

```python
df_no_duplicates.info()
```

```python
# Replace values in the 'BusinessTravel' column
df_no_duplicates['BusinessTravel'] = df_no_duplicates['BusinessTravel'].str.replace('TravelRarely', 'Travel_Rarely')
```

```python
df_no_duplicates['PerformanceRating'].unique()
```

```python
# Bar plot for Attrition distribution
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='Attrition', data=df_no_duplicates, palette='Set1')
plt.title('Attrition Distribution')

# Annotate the bar plot with counts
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

```python
# Explore the relationship between Attrition and Age using a box plot
plt.figure(figsize=(12, 8))
ax = sns.boxplot(x='Attrition', y='Age', data=df_no_duplicates, palette='Set2')
plt.title('Attrition vs. Age')

# Annotate the box plot with median values
medians = df_no_duplicates.groupby('Attrition')['Age'].median().values
median_labels = [str(int(val)) for val in medians]
pos = range(len(medians))
for tick, label in zip(pos, ax.get_xticklabels()):
    ax.text(pos[tick], medians[tick] + 0.5, median_labels[tick], 
            horizontalalignment='center', color='w', weight='bold')

plt.show()
```
